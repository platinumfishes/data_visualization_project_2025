## 1. Availability vs Access Mismatch

While broadband infrastructure may be present in many counties, it does not guarantee actual use or access by residents.

In this section, we compare fixed broadband availability (using FCC Form 477 tiered connection speeds) with household broadband adoption (from the American Community Survey).

This mismatch is important: it reveals systemic barriers such as affordability, digital literacy, or trust that prevent residents from using available infrastructure.

```python
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

# Output path for assets
assets_path = Path("assets/insights")
assets_path.mkdir(parents=True, exist_ok=True)

# === Section 1 Redo: Access vs Availability, Professional Style ===
plt.figure(figsize=(10, 7))

sns.set(style="whitegrid", font_scale=1.1)
scatter = sns.scatterplot(
    data=merged_df,
    x="connections_25mbps_per_1000",
    y="broadband_access_rate",
    hue="state_abbr",
    palette="tab20",
    alpha=0.6,
    edgecolor="w",
    linewidth=0.5,
    legend=False
)

plt.title("Broadband Access vs. Infrastructure Availability by County", fontsize=16)
plt.xlabel("Residential Connections ≥25 Mbps per 1,000 Housing Units", fontsize=13)
plt.ylabel("Broadband Access Rate (ACS)", fontsize=13)
plt.xlim(0, 1000)
plt.ylim(0, 1.05)
plt.tight_layout()

# Save professionally
scatter_path = assets_path / "availability_vs_access_scatter.png"
plt.savefig(scatter_path, dpi=300)
plt.show()
```

**Key Insight:**
Some counties show very high infrastructure availability — over 700 connections per 1,000 housing units — yet actual broadband access rates are below 60%.

This suggests that **non-technical barriers** (like cost, education, or awareness) are preventing residents from subscribing, despite service availability.

These counties may benefit more from **adoption-focused policy**, rather than additional infrastructure expansion.
